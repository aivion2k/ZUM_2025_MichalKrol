```python
import sys
from pathlib import Path

PROJECT_ROOT = Path("..").resolve()
sys.path.append(str(PROJECT_ROOT))
sys.path.append(str(PROJECT_ROOT / "src"))
```

```python
import pytorch_lightning as pl
import numpy as np
import matplotlib.pyplot as plt

from src.data.datamodule import MNISTDataModule
from src.modules.pca_subspace import PCASubspaceClassifier

dm = MNISTDataModule(for_transformer=False, batch_size=256, num_workers=4)
model = PCASubspaceClassifier(k=26)

trainer = pl.Trainer(accelerator="gpu", max_epochs=1, logger=False, enable_checkpointing=False)
trainer.fit(model, datamodule=dm)
trainer.validate(model, datamodule=dm)
trainer.test(model, datamodule=dm)
```

```python
accs = []
ks = list(range(1, 31))

for k in ks:
    m = PCASubspaceClassifier(k=k)
    t = pl.Trainer(accelerator="cpu", max_epochs=1, logger=False, enable_checkpointing=False)
    t.fit(m, datamodule=dm)
    res = t.test(m, datamodule=dm, verbose=False)[0]
    accs.append(res["test_acc"])

plt.plot(ks, accs)
plt.xlabel("k")
plt.ylabel("test_acc")
plt.title("PCA subspace: accuracy vs k")
plt.show()
```

```python
import torch
import matplotlib.pyplot as plt

@torch.no_grad()
def reconstruct(model, x, y):
    # x: [B,1,28,28], y: [B]
    X = x.view(x.size(0), -1)
    Xhat = torch.empty_like(X)
    for c in range(model.num_classes):
        idx = (y == c)
        if idx.any():
            mu = model.class_means[c]
            Uk = model.class_Uk[c]
            Z = X[idx] - mu
            Y = Z @ Uk
            Xhat[idx] = Y @ Uk.T + mu
    return Xhat.view(-1, 1, 28, 28)

# weź 100 pierwszych z test loadera
dm.setup()
xb, yb = next(iter(dm.test_dataloader()))
xb100, yb100 = xb[:100], yb[:100]
xh = reconstruct(model, xb100, yb100)

plt.figure(figsize=(8,8))
for i in range(100):
    plt.subplot(10,10,i+1)
    plt.imshow(xh[i,0].cpu(), cmap="gray")
    plt.axis("off")
plt.suptitle("Rekonstrukcje PCA(k=26) – 100 test")
plt.show()
```
